```python
# Import Library
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk mengubah pixel RGB menjadi grayscale
def convert_to_grayscale(pixel):
    r, g, b = pixel
    gray = int(0.299 * r + 0.587 * g + 0.114 * b)
    return gray
```

```python
# Membuka gambar dan mengonversinya ke mode RGB
gambar = Image.open('bunga.jpeg') 
gambar = gambar.convert('RGB')
width, height = gambar.size
```

```python
# Menampilkan gambar asli
plt.imshow(gambar)
plt.axis('off')
plt.title('Gambar Asli')
plt.show()
```

```python
# Membuat matriks kosong untuk menyimpan nilai grayscale
grayscale_matrix = []
```

```python
# Loop untuk mengakses setiap pixel gambar
for y in range(height):
    row = []
    for x in range(width):
        pixel = gambar.getpixel((x, y))
        index = convert_to_grayscale(pixel)
        row.append(index)
    grayscale_matrix.append(row)
```

```python
# Menampilkan gambar hasil konversi ke grayscale
plt.imshow(grayscale_matrix, cmap='gray')
plt.axis('off') 
plt.title('Hasil Konversi')
plt.show()
```
