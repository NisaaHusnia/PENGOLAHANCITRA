```python
# Import Library
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
# Palet warna untuk pencocokan indeks warna
palette = {
    0: (0, 0, 0),       # Hitam
    1: (255, 0, 0),     # Merah
    2: (0, 255, 0),     # Hijau
    3: (0, 0, 255)      # Biru
}
```

```python
# Fungsi untuk mencari indeks warna terdekat dari palet
def find_closest_color_index(pixel):
    closest_index = None
    min_distance = float('inf')

    for index in palette:
        r1, g1, b1 = pixel
        r2, g2, b2 = palette[index]

        # Rumus Euclidean distance untuk menghitung jarak antar warna
        distance = math.sqrt((r1 - r2)**2 + (g1 - g2)**2 + (b1 - b2)**2)

        # Jika jarak lebih kecil dari sebelumnya, perbarui indeks terdekat
        if distance < min_distance:
            min_distance = distance
            closest_index = index

    return closest_index
```

```python
# Buka gambar dan konversi ke RGB
gambar = Image.open('kucing.jpeg')
gambar = gambar.convert('RGB') 
width, height = gambar.size
```

```python
# Matriks untuk menyimpan hasil konversi terindeks
terindeks_matrix = []
```

```python
# Proses konversi setiap pixel ke indeks palet terdekat
for y in range(height):
    row = []
    for x in range(width):
        pixel = gambar.getpixel((x, y))
        index = find_closest_color_index(pixel)
        row.append(index)
    terindeks_matrix.append(row)
```

```python
# Buat gambar baru dari hasil matriks terindeks
result_img = Image.new('RGB', (width, height))
```

```python
# Ubah setiap indeks menjadi warna sesuai palet dan terapkan ke gambar baru
for y in range(height):
    for x in range(width):
        color = palette[terindeks_matrix[y][x]]
        result_img.putpixel((x, y), color)
```

```python
# Tampilkan gambar asli
plt.imshow(gambar)
plt.axis('off')
plt.title('Gambar Asli')
plt.show()
```

```python
# Tampilkan gambar hasil konversi warna terindeks
plt.imshow(result_img)
plt.axis('off')
plt.title('Hasil Konversi Terindeks')
plt.show()
```
